```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plotting(arr, labels, title=""):
    fig = plt.figure()
    plt.title(title)
    plt.ylabel("error")
    plt.xlabel("epoch")
    r, c = arr.shape
    for _ in range(c):
        plt.plot(np.arange(1,len(arr[:, _])+1), arr[:, _], label=labels[_])
    plt.legend()
```

```python
samples = 50
err = pd.read_csv(f"../model_{samples}/error.csv")
err.columns = [ "train_err", "val_err"]
for col in err.columns:
    err[col] = err[col].astype(float)

err.replace([np.inf, -np.inf], np.nan, inplace=True)
err.dropna(inplace=True)
```

```python
plotting(err.to_numpy(), err.columns)
```
