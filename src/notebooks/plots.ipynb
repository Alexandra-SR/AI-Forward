```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plotting(arr, labels, title="", ylim=10):
    fig = plt.figure()
    plt.title(title)
    plt.ylabel("error")
    plt.xlabel("epoch")
    r, c = arr.shape
    for _ in range(c):
        plt.plot(np.arange(1,len(arr[:, _])+1), arr[:, _], label=labels[_])
    plt.ylim(0, ylim)
    plt.legend()
```

```python

def show_plot(id):
    err = pd.read_csv(f"../model_{id}/error.csv")
    err.columns = [ "train_err", "val_err"]
    for col in err.columns:
        err[col] = err[col].astype(float)

    err.replace([np.inf, -np.inf], np.nan, inplace=True)
    err.dropna(inplace=True)
    plotting(err.to_numpy(), err.columns, "n = "+id)
```

```python
ids = ["13", "50", "100", "100_3_layers", "100_5_layers"]
for _ in ids:
    show_plot(_)
```

```python
err = pd.read_csv(f"../tumores/error.csv")
err.columns = [ "train_err", "val_err"]
for col in err.columns:
    err[col] = err[col].astype(float)

err.replace([np.inf, -np.inf], np.nan, inplace=True)
err.dropna(inplace=True)
plotting(err.to_numpy(), err.columns, "dataset_prueba", 1)
```
